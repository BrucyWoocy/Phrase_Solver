```python
def word_search(dict, length, pst_1, ltr_1, pst_2, ltr_2):
    list_A=[]
    list_B=[]
    list_C=[]
    pst_1=pst_1-1
    if pst_2!="":
        pst_2=pst_2-1

    for x in dict:
        a = len(x)
        if a == length:
            list_A.append(x)
    
    for x in list_A:
        if x[pst_1] == ltr_1:
         list_B.append(x)
    
    if ltr_2 != "":
        for x in list_B:
            if x[pst_2] == ltr_2:
                list_C.append(x)
        LIST.append(list_C)
    else:
        LIST.append(list_B)
```
